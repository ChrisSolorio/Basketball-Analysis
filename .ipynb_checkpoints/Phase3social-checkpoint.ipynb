```python
df2019 = pd.read_csv("Partially_cleaned_2019_set.csv")
tdata = pd.read_csv("twitterData.csv")
```

Now to understand how 'good' a player is and their impact and presence of social media, lets look at a subset of active players in 2019 with their stats and social media accounts. 

Lets first clean the data a bit to make it more useable for data analysis.

```python
df2019.drop_duplicates(subset='Player', keep='first', inplace=True)
```

```python
df2019 = df2019.set_index("Player")
tdata = tdata.set_index("player_name")
tdata.head(n=10)
```

```python
social2019 = tdata.merge(df2019, left_index=True, right_index=True)
```

```python
social2019.loc["LeBron James"]
```

Due to the outlier known as Lebron James, we can cannot visualize our data effectively. Lets remove a few outliers and try with under 1,000,000 followers

Through comparing the points per game of players versus their social media followers for all those who have under 1,000,000 followers, we can see a general linear trend where a hiegher ppg on average means more followers

```python
cols = list(underMil.columns.values)[8:]

correlations = pd.DataFrame(data = {'category': [], 'corr value': []})

for i in range(len(cols)):
    data1 = social2019['followers_count']
    data2 = social2019[cols[i]]
    corr, _ = spearmanr(data1, data2)
    correlations.loc[i] = [str(cols[i])] + [corr]

correlations
    
    
    
```

```python
underMil.plot.scatter(x='fga_per_g',y='followers_count', c='DarkBlue')
```

```python
followerstest = (underMil['followers_count'])
fgapgtest = (underMil['fga_per_g'])




from sklearn.model_selection import KFold


kfold = KFold(20, True, 1)
count = 0

arr = []

for train, test in kfold.split(followerstest, fgapgtest):
    if(count < 2):
        print('train: %s, test: %s' % (len(followerstest[train]), len(followerstest[test])))
        count += 1;
    arr.append(pd.DataFrame(data = {'fgapgtest': fgapgtest[train], 
                                    'followerstest': followerstest[train]}))


        
    
```

```python
from sklearn.linear_model import LinearRegression

log_price_model = LinearRegression()

avgcoef = 0
avgintc = 0


for testset in arr:
    log_price_model.fit(X=testset[['fgapgtest']], y=testset['followerstest'])
    print(log_price_model.coef_,log_price_model.intercept_)
    avgcoef += log_price_model.coef_[0]
    avgintc += log_price_model.intercept_
    
avgcoef = avgcoef/len(arr)
avgintc = avgintc/len(arr)
```

```python
print(avgcoef,avgintc)
```

```python

x_train = np.linspace(0, 22, num = 100)

(avgcoef*(x_train) +  avgintc)

y_new = pd.Series((avgcoef*(x_train) +  avgintc), x_train, (avgcoef*(x_train) +  avgintc))
#underMil.plot.scatter(x = 'fga_per_g', y = 'followers_count')
y_new.plot.line()
```
